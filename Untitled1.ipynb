```python
fuhao = input('请输入一个符号：')
hangshu = int(input('请输入一个行数：'))
def lingxing(h,f):
    for i in range(h):
        print(' '*(h-i),(f+' ')*i)
    for i in range(h):
        print(' '*(i),(f+' ')*(h-i))
    return  ' '
print(lingxing(hangshu,fuhao))
```

```python
# 非递归算法
def jiecheng_1(n):
    my_sum = 1
    for i in range(n):
        my_sum = my_sum * (i+1)
    return my_sum
    
def qiuhe_1(m):
    my_total = 0
    for i in range(m):
        my_total = my_total + jiecheng_1(i+1)
    return my_total
print(qiuhe_1(int(input('请输入一个整数：'))))        
```

```python
# 递归算法
def jiecheng_2(n):
    
    if n == 1:
        return 1
    else:
        return  n*jiecheng_2(n-1)

def qiuhe(n):
    if n == 1:
        return jiecheng_2(1)
    else:
        return jiecheng_2(n)+ qiuhe(n-1)
print(qiuhe(int(input('请输入一个数字'))))
```
